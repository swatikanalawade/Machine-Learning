```python
import pandas as pd
df = pd.read_csv("Ice_cream.csv")
```

```python
df.head()
x= df.iloc[:,:1]
y=df.iloc[:,-1]
```

```python
x=x.to_numpy()
y=y.to_numpy().reshape(49,1)
x.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
#creating polynomial feature
poly = PolynomialFeatures(degree=2)
fit_poly = poly.fit_transform(x)
fit_poly
```

```python
#fit the model
l = LinearRegression()
l.fit(fit_poly,y)
```

```python
l.coef_
```

```python
l.intercept_
```

```python
y_pred = l.predict(fit_poly)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='red')
plt.plot(x, y_pred, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('polynomial Regression')
plt.show()
```
